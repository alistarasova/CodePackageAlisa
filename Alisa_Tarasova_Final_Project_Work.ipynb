**Anonymous data was collected from Pratt Institute students during the day of ComD Thesis presentation. Same people were asked to choose emotion they feel at the moment, during different time of the day (9am, 12pm, 2pm, 3pm, 5pm, 7pm).  They were given the list of emotions, consisting of: "Excitement, Happiness, Delight, Enthusiasm, Joy, Tiredness, Nervousness, Sadness, Fear". Students were asked before each class and during the break, if data is missing - student didn't had class or break at that time.**

**Data was collected and writen in Google.XL. Data was then cleaned and modified in OpenReFine and saved in csv format.**

*Person 1 - Green, Purple, Red, Orange, Yellow. Circular shapes*

```python
from ColabTurtlePlus.Turtle import *
clearscreen()
setup(800,800)
import turtle
import random

person1 = Turtle()

person1.speed(10)


for i in range(180):
    person1.color("green")
    person1.forward(100)
    person1.right(30)
    person1.color("purple")
    person1.forward(20)
    person1.left(60)
    person1.color("red")
    person1.forward(50)
    person1.right(30)

    person1.color("orange")
    person1.forward(90)
    person1.right(40)
    person1.color("yellow")
    person1.forward(70)
    person1.left(20)

    person1.penup()
    person1.setposition(0, 0)
    person1.pendown()

    person1.right(2)
    person1.pendown()
```
